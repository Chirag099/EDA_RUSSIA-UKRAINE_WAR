#Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df1 = pd.read_csv("russia_losses_equipment.csv")
df2 = pd.read_csv("russia_losses_personnel.csv")
```

```python
fig = px.line(df1,x='date',y=['drone','aircraft','helicopter'],template='ggplot2',title='Russian Air Equipment Loss')
fig.show()
```

```python
names = ['aircraft','drone','helicopter']
values = [df1['aircraft'].max(),df1['drone'].max(),df1['helicopter'].max()]
fig = px.pie(names=names,values=values,hole=.8,template='ggplot2')
fig.show()
```

```python
fig = px.line(df1,x='date',y=['vehicles and fuel tanks','tank','fuel tank'],template='ggplot2',title='Russian Fuel Equipment Losses')
fig.show()
```

```python
names = ['vehicles and fuel tanks','tank','fuel tank']
values = [df1['tank'].max(),df1['fuel tank'].max(),df1['vehicles and fuel tanks'].max()]
fig = px.pie(names=names,values=values,hole=.8,template='ggplot2')
fig.show()
```

```python
fig = px.line(df1,x='date',y=['APC','military auto','field artillery','MRL','cruise missiles'],template='ggplot2',title='Russian Militay Equipment Losses')
fig.show()
```

```python
names = ['APC','military auto','field artillery','MRL','cruise missiles']
values = [df1['APC'].max(),df1['military auto'].max(),df1['field artillery'].max(),df1['MRL'].max(),df1['cruise missiles'].max()]
fig = px.pie(names=names,values=values,hole=.8,template='ggplot2')
fig.show()
```

```python
fig = px.line(df1,x='date',y=['special equipment'],template='ggplot2',title='Russian Special Equipment Losses')
fig.update_traces(showlegend = False)
fig.show()
```

```python
fig = px.line(df1,x='date',y=['naval ship'],template='ggplot2',title='Russian Naval Equipment Losses')
fig.update_traces(showlegend=False)
fig.show()
```

```python
px.bar(df1.groupby(['greatest losses direction'],as_index=False)[['tank']].count().sort_values(by='tank',ascending=False).head(15),x='greatest losses direction',y='tank',title='Cities Which Reported More Losses')
```

```python
fig = px.line(df2,x='date',y='personnel',template='plotly_dark',title='Russian Military Personnel Losses')
fig.update_layout(font_color = 'crimson')
```

From the Analysis, We get that most equipment losses took place in Donetsk city. Ukraine has lost much of its territory to Russia. On the other hand, Russia is also suffering much as it is having a lot of impact on its economy.
